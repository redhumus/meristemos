```python
#Aplicando métodos de normalización originales según explicación: https://www.youtube.com/watch?v=ro2UjY8dLWw
#Simple feature scaling
df["Dt_del10mi"] = df["Dt_del10mi"]/df["Dt_del10mi"].max() 
#Min-max
df["Mtasa_mig1"] = df["Mtasa_mig1"]-df["Mtasa_mig1"].min()/df["Mtasa_mig1"].max()-df["Mtasa_mig1"].min()
#Z-Score
df["densidad_2"] = df["densidad_2"]-df["densidad_2"].mean()/df["densidad_2"].std()

#Repetimos simple
df["tasa_ata1"] = df["tasa_ata1"]/df["tasa_ata1"].max()
```

# Aplicamos los métodos modificados para evitar negativos en densidad

```python
#Aplicando métodos de normalización modificados. Se 
#Simple feature scaling
df["Dt_del10mi"] = df["Dt_del10mi"]/df["Dt_del10mi"].max() 
#Min-max
df["Mtasa_mig1"] = df["Mtasa_mig1"]/df["Mtasa_mig1"].max()
#Z-Score
df["densidad_2"] = df["densidad_2"]/df["densidad_2"].max()

#Repetimos simple
df["tasa_ata1"] = df["tasa_ata1"]/df["tasa_ata1"].max()
```
