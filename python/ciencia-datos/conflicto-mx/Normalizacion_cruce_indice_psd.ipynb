```python
df[["Dt_del10mi", "Mtasa_mig1", "densidad_2", "tasa_ata1", "DRRIAlt"]].describe()
```

```python
#Simple feature scaling

df["Dt_del10mi"] = df["Dt_del10mi"]/df["Dt_del10mi"].max()
```

```python
#Min-Max
df["Mtasa_mig1"] = df["Mtasa_mig1"]-df["Mtasa_mig1"].min()/df["Mtasa_mig1"].max()-df["Mtasa_mig1"].min()
```

```python

df["densidad_2"] = df["densidad_2"]-df["densidad_2"].mean()/df["densidad_2"].std()

```

```python
# Nombre del campo para el que deseas calcular la media y la mediana
campo = "nombre_del_campo"

# Calcular la media y la mediana
media = df[campo].mean()
mediana = df[campo].median()

# Mostrar los resultados
print(f"Media de {campo}: {media}")
print(f"Mediana de {campo}: {mediana}")
```

```python
import pandas as pd

# Supongamos que ya tienes tu DataFrame df con las variables
variables = ["Dt_del10mi", "Mtasa_mig1", "densidad_2", "tasa_ata1", "DRRIAlt"]
description = df[variables].describe()

# Calcula la mediana y crea un DataFrame para la fila de la mediana
median_row = pd.DataFrame(df[variables].median(), columns=['median']).transpose()

# Concatena el DataFrame de la mediana con la descripción
description_with_median = pd.concat([description, median_row])

print(description_with_median)

```

# Transformación logarítmica

```python
import numpy as np
import pandas as pd

#  DataFrame llamado df y las variables del índice en la lista "variables"
variables = ["Dt_del10mi", "Mtasa_mig1", "densidad_2", "tasa_ata1", "DRRIAlt"]

# Recorre la lista de variables y crea nuevos campos logaritmizados
for variable in variables:
    new_field_name = f"l_{variable}"
    df[new_field_name] = np.log(df[variable] + 1)  # Añade 1 para evitar log(0)

# Verifica el DataFrame con los nuevos campos logaritmizados
print(df)

```

```python
# Guarda el DataFrame modificado en un nuevo archivo CSV
new_csv_filename = "/home/omnibus/Documents/QGISS_Cartografia/Proyectos/criminal_mx/indice_psd/cruce_trans_log.csv"
df.to_csv(new_csv_filename, index=False)  # El parámetro index=False evita guardar el índice como columna

print(f"DataFrame modificado guardado en {new_csv_filename}")
```

# Verificación media / mediana

Podemos ver que los datos se normalizaron en gran medida

```python
variables = ["l_Dt_del10mi", "l_Mtasa_mig1", "l_densidad_2", "l_tasa_ata1", "l_DRRIAlt", "indice"]

description = df[variables].describe()

# Calcula la mediana y crea un DataFrame para la fila de la mediana
median_row = pd.DataFrame(df[variables].median(), columns=['median']).transpose()

# Concatena el DataFrame de la mediana con la descripción
description_with_median = pd.concat([description, median_row])

print(description_with_median)
```

# Creando el índice

```python
import numpy as np
import pandas as pd

#DataFrame llamado df con las variables y los campos logaritmizados

# Calcula el valor del índice para cada registro
df["indice"] = (df["l_Dt_del10mi"] + df["l_Mtasa_mig1"] + df["l_densidad_2"] + df["l_tasa_ata1"] + df["l_DRRIAlt"]) / 5

# Muestra el DataFrame con el nuevo campo "indice"
print(df)
```
